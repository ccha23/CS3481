Load the iris dataset.

```
library(datasets)
data(iris)
summary(iris)
```

Plot the iris dataset.

```
plot(iris)
```

### CART

Install the library for CART.

```
if(!require('rpart'))
   install.packages('rpart')
library(rpart)
```

Construct the classification tree.

```
CART <- rpart(Species ~ ., method="class", data=iris)
# display the results
printcp(CART) 
# Plot the tree
plot(CART, uniform=TRUE, main="Classification Tree")
text(CART, use.n=TRUE, all=TRUE, cex=.8)
```
